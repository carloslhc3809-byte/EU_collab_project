# Data preparation

```python
from pathlib import Path
import pandas as pd

# Detect project root (go up until "data" folder is found)
current = Path().resolve()

for parent in [current] + list(current.parents):
    if (parent / "data").exists():
        PROJECT_ROOT = parent
        break
else:
    raise RuntimeError("Could not find project root (no 'data' folder found)")

# Build paths
REAL_PATH = PROJECT_ROOT / "data/raw/EU_C12N15_data.csv"
SAMPLE_PATH = PROJECT_ROOT / "data/sample.csv"

# Select dataset
if REAL_PATH.exists():
    print("Using full dataset")
    DATA_PATH = REAL_PATH
elif SAMPLE_PATH.exists():
    print("Using sample dataset")
    DATA_PATH = SAMPLE_PATH
else:
    raise FileNotFoundError(
        f"No dataset found.\nChecked:\n- {REAL_PATH}\n- {SAMPLE_PATH}"
    )

# Load
raw_data = pd.read_csv(DATA_PATH)

# Process
raw_data["family_id"] = raw_data["Simple Family Members"].astype(str).str.split(";;").str[0]
raw_data = raw_data.drop_duplicates(subset="family_id")

print(f"Loaded {len(raw_data)} rows from {DATA_PATH}")
```

```python
import pandas as pd
import ast

def parse_list_column(x):
    if pd.isna(x):
        return []
    if isinstance(x, list):
        return x
    try:
        return ast.literal_eval(x)
    except:
        return [i.strip() for i in str(x).split(";")]
```

```python
df = raw_data[raw_data["Publication Year"] < 2025].copy()
df["Applicants"] = df["Applicants"].apply(parse_list_column)
df["Inventors"] = df["Inventors"].apply(parse_list_column)
df["CPC Classifications"] = df["CPC Classifications"].apply(parse_list_column)
df["Priority Numbers"] = df["Priority Numbers"].apply(parse_list_column)
```

```python
from torch_geometric.data import HeteroData
import torch

def build_pyg_hetero_graph(df):

    data = HeteroData()

    # -----------------------
    # Create node mappings
    # -----------------------

    applicants = set()
    inventors = set()
    patents = set()
    cpcs = set()

    for _, row in df.iterrows():

        pats = row["Priority Numbers"] or [row["Lens ID"]]

        for p in pats:
            patents.add(p)

        applicants.update(row["Applicants"])
        inventors.update(row["Inventors"])
        cpcs.update(row["CPC Classifications"])

    # index mapping
    app_map = {a:i for i,a in enumerate(applicants)}
    inv_map = {i:i2 for i2,i in enumerate(inventors)}
    pat_map = {p:i for i,p in enumerate(patents)}
    cpc_map = {c:i for i,c in enumerate(cpcs)}

    # Build edges
    app_pat_edges = []
    inv_pat_edges = []
    cpc_pat_edges = []

    for _, row in df.iterrows():

        pats = row["Priority Numbers"] or [row["Lens ID"]]

        for p in pats:

            if p not in pat_map:
                continue

            p_id = pat_map[p]

            for a in row["Applicants"]:
                if a in app_map:
                    app_pat_edges.append([app_map[a], p_id])

            for i in row["Inventors"]:
                if i in inv_map:
                    inv_pat_edges.append([inv_map[i], p_id])

            for c in row["CPC Classifications"]:
                if c in cpc_map:
                    cpc_pat_edges.append([cpc_map[c], p_id])

    # Convert to tensors
    def to_edge_index(edges):
        if len(edges) == 0:
            return torch.empty((2,0), dtype=torch.long)
        return torch.tensor(edges, dtype=torch.long).t().contiguous()

    data["applicant","files","patent"].edge_index = to_edge_index(app_pat_edges)
    data["patent","rev_files","applicant"].edge_index = to_edge_index([[j,i] for i,j in app_pat_edges])

    data["inventor","invented","patent"].edge_index = to_edge_index(inv_pat_edges)
    data["patent","rev_invented","inventor"].edge_index = to_edge_index([[j,i] for i,j in inv_pat_edges])

    data["cpc","in_patent","patent"].edge_index = to_edge_index(cpc_pat_edges)
    data["patent","rev_in_patent","cpc"].edge_index = to_edge_index([[j,i] for i,j in cpc_pat_edges])

    # Node counts
    data["applicant"].num_nodes = len(app_map)
    data["inventor"].num_nodes = len(inv_map)
    data["patent"].num_nodes = len(pat_map)
    data["cpc"].num_nodes = len(cpc_map)

    return data
```

```python
data_by_year = {}

years = sorted(df["Publication Year"].dropna().unique())

for year in years:

    df_year = df[df["Publication Year"] == year]  

    data_year = build_pyg_hetero_graph(df_year)

    data_by_year[year] = data_year
```

```python
import pandas as pd
import ast
import torch
from torch_geometric.data import HeteroData

# ---------- helpers ----------
def parse_list(x):
    if pd.isna(x): return []
    try: return ast.literal_eval(x)
    except: return [i.strip() for i in str(x).split(";")]

df = raw_data[raw_data["Publication Year"] < 2025].copy()
for col in ["Applicants","Inventors","CPC Classifications","Priority Numbers"]:
    df[col] = df[col].apply(parse_list)

# ---------- graph builder ----------
def build_graph(df):

    data = HeteroData()

    apps = set(); invs = set(); pats = set(); cpcs = set()

    for _, r in df.iterrows():
        ps = r["Priority Numbers"] or [r["Lens ID"]]
        pats.update(ps)
        apps.update(r["Applicants"])
        invs.update(r["Inventors"])
        cpcs.update(r["CPC Classifications"])

    app_map = {a:i for i,a in enumerate(apps)}
    inv_map = {i:i2 for i2,i in enumerate(invs)}
    pat_map = {p:i for i,p in enumerate(pats)}
    cpc_map = {c:i for i,c in enumerate(cpcs)}

    ap=[]; ip=[]; cp=[]; collab=[]

    for _, r in df.iterrows():

        ps = r["Priority Numbers"] or [r["Lens ID"]]

        # applicant–patent
        for p in ps:
            if p not in pat_map: continue
            pid = pat_map[p]

            for a in r["Applicants"]:
                if a in app_map:
                    ap.append([app_map[a], pid])

            for i in r["Inventors"]:
                if i in inv_map:
                    ip.append([inv_map[i], pid])

            for c in r["CPC Classifications"]:
                if c in cpc_map:
                    cp.append([cpc_map[c], pid])

        # ---------- Collab edges INSIDE function ----------
        apps_list = r["Applicants"]

        for i in range(len(apps_list)):
            for j in range(i+1, len(apps_list)):

                a_i = apps_list[i]
                a_j = apps_list[j]

                if a_i in app_map and a_j in app_map:

                    i_idx = app_map[a_i]
                    j_idx = app_map[a_j]

                    collab.append([i_idx, j_idx])
                    collab.append([j_idx, i_idx])

    # ---------- tensor helper ----------
    def to_tensor(e):
        return torch.tensor(e, dtype=torch.long).t().contiguous() if e else torch.empty((2,0), dtype=torch.long)

    # ---------- assign edges ----------
    data["applicant","files","patent"].edge_index = to_tensor(ap)
    data["patent","rev_files","applicant"].edge_index = to_tensor([[j,i] for i,j in ap])

    data["inventor","invented","patent"].edge_index = to_tensor(ip)
    data["patent","rev_invented","inventor"].edge_index = to_tensor([[j,i] for i,j in ip])

    data["cpc","in_patent","patent"].edge_index = to_tensor(cp)
    data["patent","rev_in_patent","cpc"].edge_index = to_tensor([[j,i] for i,j in cp])

    # Collab edge
    data["applicant","collab","applicant"].edge_index = to_tensor(collab)

    data["applicant"].num_nodes = len(app_map)

    return data, app_map


# ---------- temporal graphs ----------
data_by_year, app_maps = {}, {}

for y in sorted(df["Publication Year"].dropna().unique()):
    d = df[df["Publication Year"] <= y]
    data_by_year[y], app_maps[y] = build_graph(d)
```

```python
from torch_geometric.nn import MetaPath2Vec
import torch

# ---------- metapaths ----------
metapaths = {
    "APA":[("applicant","files","patent"),("patent","rev_files","applicant")],
    "API":[("applicant","files","patent"),("patent","rev_invented","inventor"),
           ("inventor","invented","patent"),("patent","rev_files","applicant")],
    "APC":[("applicant","files","patent"),("patent","rev_in_patent","cpc"),
           ("cpc","in_patent","patent"),("patent","rev_files","applicant")]
}

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# ---------- MetaPath training ----------
def train_mp(data, mp):

    try:
        model = MetaPath2Vec(
            data.edge_index_dict,
            embedding_dim=32,
            metapath=mp,
            walk_length=30,
            context_size=5,
            walks_per_node=3,
            num_negative_samples=5,
            sparse=True
        ).to(device)

    except Exception:
        return torch.zeros(data["applicant"].num_nodes, 32)

    opt = torch.optim.SparseAdam(model.parameters(), lr=0.01)
    loader = model.loader(batch_size=128, shuffle=True)

    model.train()

    for _ in range(2):
        for pos_rw, neg_rw in loader:

            pos_rw = pos_rw.to(device)
            neg_rw = neg_rw.to(device)

            opt.zero_grad()
            loss = model.loss(pos_rw, neg_rw)
            loss.backward()
            opt.step()

    model.eval()

    with torch.no_grad():
        emb = model("applicant").cpu()

    return emb


# ---------- embeddings per year ----------
emb_by_year = {}

for y, data in data_by_year.items():

    embs = []

    for name, mp in metapaths.items():
        emb = train_mp(data, mp)
        embs.append(emb)

    emb_by_year[y] = torch.cat(embs, dim=1)


# ---------- Global mapping ----------
global_apps = {}

for app_map in app_maps.values():
    for app in app_map:
        if app not in global_apps:
            global_apps[app] = len(global_apps)


# ---------- temporal tensor ----------
years_sorted = sorted(emb_by_year.keys())

T = len(years_sorted)
N = len(global_apps)
D = next(iter(emb_by_year.values())).shape[1]

X = torch.zeros(T, N, D)

# ---------- Alignment ----------
for t, y in enumerate(years_sorted):

    z = emb_by_year[y]
    app_map = app_maps[y]

    for app, local_idx in app_map.items():

        if app in global_apps:
            global_idx = global_apps[app]
            X[t, global_idx] = z[local_idx]
```

# Baseline Model

```python
import torch
import torch.nn as nn
import random
import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score, average_precision_score
from torch_geometric.utils import negative_sampling

seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)

# Edge mapping
def build_global_edge_index(local_edge_index, app_map, global_map):

    inv_app_map = {v:k for k,v in app_map.items()}
    edges = []

    for src, dst in local_edge_index.t().tolist():

        if src not in inv_app_map or dst not in inv_app_map:
            continue

        a_src = inv_app_map[src]
        a_dst = inv_app_map[dst]

        if a_src in global_map and a_dst in global_map:
            edges.append([global_map[a_src], global_map[a_dst]])

    if len(edges) == 0:
        return torch.empty((2,0), dtype=torch.long)

    return torch.tensor(edges, dtype=torch.long).t().contiguous()


# Model (baseline)
class Model(nn.Module):
    def __init__(self, in_dim, hid):
        super().__init__()

        self.gru = nn.GRU(in_dim, hid)
        self.mlp = nn.Sequential(
            nn.Linear(hid * 4, hid),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hid, 1)
        )

    def forward(self, X, edge_index):

        h, _ = self.gru(X)      # [T,N,H]
        z = h.mean(dim=0)       # instead of h[-1]

        if edge_index.size(1) == 0:
            return torch.tensor([]), z

        src, dst = edge_index
        s = z[src]
        d = z[dst]
        feat = torch.cat([
            s,
            d,
            torch.abs(s - d),
            s * d
        ], dim=1)

        out = torch.sigmoid(self.mlp(feat)).view(-1)

        return out, z


# Temporal split
years = sorted(data_by_year.keys())

train_years = years[:-1]
test_year = years[-1]

train_T = len(train_years)

# Normalization
X_train = X[:train_T]

mean = X_train.mean(dim=(0,1), keepdim=True)
std  = X_train.std(dim=(0,1), keepdim=True) + 1e-8

X = (X - mean) / std
X_train = X[:train_T]

# Train edges 
train_edges = []

for y in train_years:

    local_edge_index = data_by_year[y]["applicant","collab","applicant"].edge_index

    global_edge = build_global_edge_index(
        local_edge_index,
        app_maps[y],
        global_apps
    )

    if global_edge.size(1) > 0:
        train_edges.append(global_edge)

train_edge_index = torch.cat(train_edges, dim=1)

# Test edges (all edges)
test_local_edge = data_by_year[test_year]["applicant","collab","applicant"].edge_index

test_edge_index = build_global_edge_index(
    test_local_edge,
    app_maps[test_year],
    global_apps
)

print("Train edges:", train_edge_index.shape)
print("Test edges:", test_edge_index.shape)

# Train model
model = Model(X_train.shape[-1], 64)
opt = torch.optim.Adam(model.parameters(), lr=5e-4)

for epoch in range(80):

    model.train()

    pos_pred, z = model(X_train.detach(), train_edge_index)

    neg_edge_index = negative_sampling(
        train_edge_index,
        num_nodes=z.size(0),
        num_neg_samples=train_edge_index.size(1)
    )

    neg_pred, _ = model(X_train.detach(), neg_edge_index)

    eps = 1e-15
    loss = -torch.log(pos_pred + eps).mean() \
           -torch.log(1 - neg_pred + eps).mean()

    opt.zero_grad()
    loss.backward()

    # stabilize GRU training
    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

    opt.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

# Evaluation
model.eval()

with torch.no_grad():

    pos_pred, z = model(X_train.detach(), test_edge_index)

    neg_edge_index = negative_sampling(
        test_edge_index,
        num_nodes=z.size(0),
        num_neg_samples=test_edge_index.size(1)
    )

    neg_pred, _ = model(X_train.detach(), neg_edge_index)

    preds = torch.cat([pos_pred, neg_pred]).cpu().numpy()
    labels = torch.cat([
        torch.ones(pos_pred.size(0)),
        torch.zeros(neg_pred.size(0))
    ]).cpu().numpy()

    auc = roc_auc_score(labels, preds)
    ap = average_precision_score(labels, preds)

print("\nFINAL MODEL PERFORMANCE:")
print(f"AUC: {auc:.4f}, AP: {ap:.4f}")
```

# Final Model

```python
import torch
import torch.nn as nn
import random
import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score, average_precision_score
from torch_geometric.utils import negative_sampling

seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)

# Edge mapping
def build_global_edge_index(local_edge_index, app_map, global_map):

    inv_app_map = {v:k for k,v in app_map.items()}
    edges = []

    for src, dst in local_edge_index.t().tolist():

        if src not in inv_app_map or dst not in inv_app_map:
            continue

        a_src = inv_app_map[src]
        a_dst = inv_app_map[dst]

        if a_src in global_map and a_dst in global_map:
            edges.append([global_map[a_src], global_map[a_dst]])

    if len(edges) == 0:
        return torch.empty((2,0), dtype=torch.long)

    return torch.tensor(edges, dtype=torch.long).t().contiguous()


# Model with temporal attention
class Model(nn.Module):
    def __init__(self, in_dim, hid):
        super().__init__()

        self.gru = nn.GRU(in_dim, hid)

        # temporal attention
        self.att_proj = nn.Linear(hid, 1)

        # MLP 
        self.mlp = nn.Sequential(
            nn.Linear(hid * 4, hid),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hid, 1)
        )

    def forward(self, X, edge_index):

        h, _ = self.gru(X)      # [T,N,H]

        # Temporal attention
        scores = torch.softmax(self.att_proj(h), dim=0)   # [T,N,1]
        z = (h * scores).sum(dim=0)                       # [N,H]

        if edge_index.size(1) == 0:
            return torch.tensor([]), z

        src, dst = edge_index

        s = z[src]
        d = z[dst]

        # interaction-aware features
        feat = torch.cat([
            s,
            d,
            torch.abs(s - d),
            s * d
        ], dim=1)

        out = torch.sigmoid(self.mlp(feat)).view(-1)

        return out, z

# Temporal split
years = sorted(data_by_year.keys())

train_years = years[:-1]
test_year = years[-1]

train_T = len(train_years)

# Normalization
X_train = X[:train_T]

mean = X_train.mean(dim=(0,1), keepdim=True)
std  = X_train.std(dim=(0,1), keepdim=True) + 1e-8

X = (X - mean) / std
X_train = X[:train_T]

# Train edges (collab edges from all train years)
train_edges = []

for y in train_years:

    local_edge_index = data_by_year[y]["applicant","collab","applicant"].edge_index

    global_edge = build_global_edge_index(
        local_edge_index,
        app_maps[y],
        global_apps
    )

    if global_edge.size(1) > 0:
        train_edges.append(global_edge)

train_edge_index = torch.cat(train_edges, dim=1)

# Test edges (all collab edges from test year)
test_local_edge = data_by_year[test_year]["applicant","collab","applicant"].edge_index

test_edge_index = build_global_edge_index(
    test_local_edge,
    app_maps[test_year],
    global_apps
)

print("Train edges:", train_edge_index.shape)
print("Test edges:", test_edge_index.shape)

# Model training
model = Model(X_train.shape[-1], 64)
opt = torch.optim.Adam(model.parameters(), lr=5e-4)

for epoch in range(40):

    model.train()

    pos_pred, z = model(X_train.detach(), train_edge_index)

    neg_edge_index = negative_sampling(
        train_edge_index,
        num_nodes=z.size(0),
        num_neg_samples=train_edge_index.size(1)
    )

    neg_pred, _ = model(X_train.detach(), neg_edge_index)

    eps = 1e-15
    loss = -torch.log(pos_pred + eps).mean() \
           -torch.log(1 - neg_pred + eps).mean()

    opt.zero_grad()
    loss.backward()

    # 🔥 stabilize GRU training
    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

    opt.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

# Evaluation on test set
model.eval()

with torch.no_grad():

    pos_pred, z = model(X_train.detach(), test_edge_index)

    neg_edge_index = negative_sampling(
        test_edge_index,
        num_nodes=z.size(0),
        num_neg_samples=test_edge_index.size(1)
    )

    neg_pred, _ = model(X_train.detach(), neg_edge_index)

    preds = torch.cat([pos_pred, neg_pred]).cpu().numpy()
    labels = torch.cat([
        torch.ones(pos_pred.size(0)),
        torch.zeros(neg_pred.size(0))
    ]).cpu().numpy()

    auc = roc_auc_score(labels, preds)
    ap = average_precision_score(labels, preds)

print("\nFINAL MODEL PERFORMANCE:")
print(f"AUC: {auc:.4f}, AP: {ap:.4f}")
```
